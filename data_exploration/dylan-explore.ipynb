```python
# imports 
# imports
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
import numpy as np
from skimage.metrics import mean_squared_error
import os

# consts
path = '/Users/dylan.geldenhuys/lifeq_work/metgrated_micah_get_Image/camera_id_9'
list_image_filenames = os.listdir('/Users/dylan.geldenhuys/lifeq_work/metgrated_micah_get_Image/camera_id_9')
```

```python
# Analysing some images 


path1 = os.path.join(path,'0d266782-80e7-44f8-8894-060da6.jpg' )
path2 = os.path.join(path,'0f95a884-3ec9-482a-ac3b-8b3d83.jpg' )
path3 = os.path.join(path,'dad77fd9-5ab9-4987-8e06-fa790a.jpg' )
path4 = os.path.join(path,'b27c9331-670d-412b-8e9e-dfbae4.jpg' )
path5 = os.path.join(path,'fed70a2e-96a4-4ba5-889b-e7a738.jpg' )
```

```python
# Utility functions

def pre_process(image):
    # converts and image to gray scale (single channel) and does min max 
    # scalling to scale the pixel values to have the same range. 
    result = image
    if (len(image.shape) == 3):
        result = rgb2gray(image)
    
    return (result - result.min()) / (result.max() - result.min())

def plot_image(image):
    plt.imshow(image)
    plt.show()

def load_image(file_name):
    return plt.imread(os.path.join(path, file_name))
```

```python
plot_image(load_image(path1))
plot_image(load_image(path2))
plot_image(load_image(path3))
plot_image(load_image(path4))
```

# Trying out a simularity function

```python
simularity_same_camera = mean_squared_error(pre_process(load_image(path1)), pre_process(load_image(path2)))
simularity_diff_camera = mean_squared_error(pre_process(load_image(path5)), pre_process(load_image(path4)))

print('MSE for same camera: ', simularity_same_camera)
print('MSE for different camera: ', simularity_diff_camera)
```

Yay you can clearly see a nice difference here.
